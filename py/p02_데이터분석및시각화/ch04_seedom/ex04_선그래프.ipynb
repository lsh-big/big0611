# 선그래프(lineplot)
- 일기예보에서 이번주 예상 기온 변화
- 시간에 따른 주가 변화량

```python
import seaborn as sns

# 1. 기본 선그래프 생성
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 42]

sns.lineplot(x=x, y=y)
```

```python
# 2. tips 데이터를 활용한 선그래프
df = sns.load_dataset('tips')
# 인원 수별 팁
sns.lineplot(x=df['size'], y=df['tip'])
```

```python
# 2. tips 데이터를 활용한 선그래프
df = sns.load_dataset('tips')
# 인원 수별 팁
# orient: 선의 방향 (x(기본값), y(역방향))
sns.lineplot(x=df['tip'], y=df['size'], orient='y')
```

```python
print('min:{}, max:{}'.format(
    min(df[df['size'] == 5]['tip']),
    max(df[df['size'] == 5]['tip'])
))
```

```python
# 3. ci(Confidence Interval: 신뢰구간) 옵션
# ci=100이면 100% 신뢰구간(=전체 범위)
# errorbar=None이면 신뢰구간 표시 안 함
# 기본값은 95%
# sns.lineplot(x=df['size'], y=df['tip'], ci=100)
sns.lineplot(x=df['size'], y=df['tip'], errorbar=('ci', 100))
```

```python
# 신뢰구간x
sns.lineplot(x=df['size'], y=df['tip'], errorbar=None)
```

```python
# 요일 별, 인원수 별 팁
sns.lineplot(x=df['size'], y=df['tip'], hue=df['day'], errorbar=None)
```

```python
# hue 속성에 여러 개의 값이 포함된 변수를 지정한 경우
sns.lineplot(x=df['day'], y=df['total_bill'], hue=df['tip'], errorbar=None)
```
