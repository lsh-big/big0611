# 회귀선(regplot)
- regplot와 scatterplot는 모두 데이터 간의 상관관계를 파악할 때 유용하다.
## tips 데이터를 활용한 regplot 실습

```python
import seaborn as sns
df = sns.load_dataset('tips')
```

```python
# 1. 기본 회귀선 그래프
sns.regplot(x=df['total_bill'], y=df['tip'])
```

```python
# 2. 회귀선만 표시하기
sns.regplot(data=df, x='total_bill', y='tip', scatter=False)
```

```python
# 3. 신뢰구간의 이해
sns.regplot(data=df, x='total_bill', y='tip', ci=None, scatter=False)
```
