### Series와 DataFrame 예시

```python
import pandas as pd

# Series 생성 : 1차원 데이터
temperatures = pd.Series([23.5, 25.1, 19.8, 30.2], index=['월', '화', '수', '목'], name='온도')
print(temperatures)
```

```python
# DataFrame 생성 (2차원 데이터)
weather_data = pd.DataFrame({
    '온도': [23.5, 25.1, 19.8, 30.2],
    '습도': [60, 65, 70, 55],
    '강수량': [0, 0, 5.2, 0]
}, index=['월', '화', '수', '목'])
print(weather_data)
```

### NumPy 와 DataFrame 비교

```python
import numpy as np

# 숫자 인덱스만 사용 가능
data_array = np.array([[23.5, 60, 0], [25.1, 65, 0], [19.8, 70, 5.2], [30.2, 55, 0]])
# 데이터 접근이 직관적이지 않음
print(data_array[0, 0])  # 첫번째 날의 온도: 23.5
```

```python
# 의미 있는 인덱스와 컬럼명 사용
print(weather_data.loc['월', '온도'])
print(weather_data['온도'])
print(weather_data.loc['수'])
```
