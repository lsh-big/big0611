```python
# 리스트 연산 -> 반복문 필요!
# 예) [1, 2, 3, 4] -> [2, 4, 6, 8]
list_data = [1, 2, 3, 4]
resutt_list = []
for item in list_data:
    resutt_list.append(item * 2)
print(resutt_list)
resutt_list
```

```python
# 배열 연산 -> 반복문 불필요
# 백터화 연산
array_data = np.array([1, 2, 3, 4])
result_array = array_data * 2  # 모든 원소에 2를 곱함
```
