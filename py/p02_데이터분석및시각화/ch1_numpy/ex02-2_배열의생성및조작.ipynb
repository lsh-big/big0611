# 최신 표준: Generator 객체 사용법

```python
import numpy as np

# 1. 최신 표준 Generator 생성
rng = np.random.default_rng()
print(rng)
```

# rng.integers(low, high, size): 정수 무작위 추출
- 내장 모듈의randint와 달리high미만(포함되지 않음)의 정수 배열을 만듭니다.size옵션에 원하는 행렬 형태(튜플)를 지정할 수 있습니다.

```python
# 1 이상 46 미만(1~45)의 숫자 중 6개를 1차원 배열로 추출
lotto = rng.integers(1, 46, size=6)
print(lotto)  # 예: [12 45  3 24 18  7]

# 0 이상 10 미만의 정수로 채워진 2행 3열짜리 배열 생성
matrix_int = rng.integers(0, 10, size=(2, 3))
print(matrix_int)
```

# rng.random(size): 0.0 이상 1.0 미만의 실수 추출M
- 원하는 형태(shape)의 난수 돗자리를 0~1 사이의 실수로 채울 때 사용합니다.

```python
# 0.0 ~ 1.0 사이의 실수로 채워진 3행 2열 배열 생성
matrix_float = rng.random((3, 2))
print(matrix_float)
```

# rng.choice(a, size, replace): 샘플 추출 (복원/비복원)
- a: 선택할 대상 배열
- size: 추출할 요소 개수
- replace: True(복원), False(비복원)

```python
items = ['Apple', 'Banana', 'Orange', 'Grape']

# 중복을 허용하여 2개 선택 (기본값: replace=True)
print(rng.choice(items, size=2))

# 중복 없이 3개 선택 (비복원 추출)
print(rng.choice(items, size=3, replace=False))
```

# rng.shuffle(x)/rng.permutation(x): 무작위 무작위 셔플
- shuffle: 원본 배열 자체를 내부에서 직접 뒤섞어버립니다. (반환값 없음)
- permutation: 원본 배열을 수정하지 않고, 새롭게 뒤섞인 배열을 반환합니다.

```python
arr = np.array([1, 2, 3, 4, 5])

shuffled = rng.permutation(arr)
print("원본:", arr)       # [1, 2, 3, 4, 5] 그대로 유지
print("셔플본:", shuffled) # 무작위 뒤섞인 배열
```

# 확률 분포 기반 난수 생성 (통계·머신러닝 필수)
## rng.normal(loc, scale, size): 정규 분포 (Gaussian)

```python
# 평균 0, 표준편차 1인 표준정규분포를 따르는 데이터 10,000개 생성
normal_data = rng.normal(loc=0.0, scale=1.0, size=10000)
print(normal_data)
```

# rng.uniform(low, high, size): 균등 분포 (Uniform)
- 지정한 범위 내의 모든 실수가 동일한 확률(균등한 빈도)로 등장하는 난수 배열을 만듭니다.

```python
# 10.0부터 20.0 사이에서 균등한 확률로 5개 추출
uniform_data = rng.uniform(10.0, 20.0, size=5)
print(uniform_data)
```

### 배열의 압축된 형식 - npz 형식
- np.savez(파일이름, 저장할_배열): 여러 개의 배열을 하나의 파일로 저장
- np.savez_compressed(파일이름, 저장할_배열): 여러 개의 배열을 압축하여 하나의 파일로 저장

```python
#### 1. np.savez() - 압축된 npy 파일 생성
```

```python
# 여러 배열 생성
arr1 = np.arange(10)
arr2 = np.arange(20).reshape(4, 5)
arr3 = np.random.rand(3, 3)
```

```python
# 여러 배열을 하나의 파일로 저장
np.savez('multiple_arrays', first_array=arr1, second_array=arr2, third_array=arr3)
```

```python
# 또는 이름 없이 저장
np.savez('arrays_no_names', arr1, arr2, arr3)
```

```python
#### 2. npz 파일 불러오기 - np.load()
```

```python
# npz 파일 불러오기
loaded_data = np.load('multiple_arrays.npz')

# 저장된 배열 이름 확인

print('117:',loaded_data.files)
```

```python
# 개별 배열 접근
print(loaded_data['first_array'])
print(loaded_data['second_array'])
print(loaded_data['third_array'])
```

```python
# 또는 변수로 할당
arr1_loaded = loaded_data['first_array']
arr2_loaded = loaded_data['second_array']
arr3_loaded = loaded_data['third_array']
```

```python
# 사용 후 파일 닫기
loaded_data.close()
```

### 압축 저장 - np.savez_compressed()

```python
# 압축하여 저장(파일 크기 절약)
np.savez_compressed('compressed_arrays', data1=arr1, data2=arr2, data3=arr3)
```
