 2. 비교 연산자
- \> 크다(초과)
- < 작다(미만)
- \>= 크거나 같다(이상)
- <= 작거나 같다(이하)
- == 같다
- != 같지않다

```python
print(a > b) 
print(a < b)
print(a == b)
print(a != b)
print(a >= b)
print(a <= b)
```

#### 3.논리연산자
- and(논리곱)
- or(논리합)
- not(논리부정)

```python
print((10 > 5) and (3 > 1))
print((10 > 5) or (3 < 1))
print(not (10 > 5))
```

```python
x = 5           #파이썬은 ++, -- 연산자 없음
x += 3
print(x)
x -= 3
x *= 3
print(x)
x /= 3
x //= 3
print(x)
x %= 3
x **= 3
print(x)
```

#### 6. 멤버십 연산자
- in : 값이 안에 있으면 True, 없으면 False
- not in : 값이 안에 없으면 True, 있으면 False

```python
fruits = ["apple", "banana", "cherry"]
print("banana" in fruits)      # 결과: True
print("orange" not in fruits)  # 결과: True
```
